```python
from matplotlib import pyplot as plt
import numpy as np

x = list(range(0,10))
y = [0.1, 0.5, -1.8, 3, 2.1, 15, 1.6, 1.8, -1, -2]

a, b, c = np.polyfit(np.asarray(x).astype(float), np.asarray(y).astype(float), 2)

linx = np.linspace(x[0], x[-1], 1000)
liny = a * linx ** 2 + b * linx + c

plt.scatter(x, y)
plt.plot(linx, liny)
plt.show()
```
